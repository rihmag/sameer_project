# Weapon Detection Training Notebook

This notebook provides the environment to train a YOLO model specifically for weapon detection using your custom dataset.

```python
# Install the ultralytics package if not already installed
!pip install ultralytics
```

```python
from ultralytics import YOLO
import os
import torch

# Check M5 MPS availability
if torch.backends.mps.is_available():
    print("✅ Apple M5 MPS GPU detected — using Metal acceleration")
else:
    print("⚠️ MPS not available — falling back to CPU")

model = YOLO('yolov8s.pt')  # nano = faster on M5

dataset_path = os.path.abspath('more_dataset/data.yaml')
if not os.path.exists(dataset_path):
    raise FileNotFoundError(
        f'Could not find {dataset_path}. '
        f'Please ensure data.yaml exists in the current directory.'
    )

results = model.train(
    data=dataset_path,
    epochs=25,          # ⬇ reduced from 100
    imgsz=416,          # ⬇ reduced from 640 (faster on M5)
    batch=16,
    name='weapon_yolo_model',

    device='mps',       # ✅ Apple Silicon — M1/M2/M3/M4/M5
    workers=4,          # M5 has good CPU cores, keep at 4
    patience=15,        # ⬇ stop earlier if no improvement
    save_period=10,

    # Optimizer
    lr0=0.01,
    lrf=0.001,
    momentum=0.937,
    weight_decay=0.0005,

    # Lighter augmentation = faster training
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    fliplr=0.5,
    flipud=0.0,         # ⬇ disabled
    mosaic=0.5,         # ⬇ reduced from 1.0
    mixup=0.0,          # ⬇ disabled
    amp=False,          # ❌ disable — MPS has issues with mixed precision
)

print("\n✅ Training Complete!")
print(f"Best model saved at: runs/detect/weapon_yolo_model/weights/best.pt")
```

```python
# 3. Validate and Export
metrics = model.val()
print(f"Mean Average Precision: {metrics.box.map}")

# Export the trained weights to the models directory for use in detect.py
# path = model.export(format='onnx') 
```
