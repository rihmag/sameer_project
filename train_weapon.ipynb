# Weapon Detection Training Notebook

This notebook provides the environment to train a YOLO model specifically for weapon detection using your custom dataset.

```python
from ultralytics import YOLO
import os

# 1. Initialize the model
# We start with a pre-trained yolov8n (nano) model for faster training
model = YOLO('yolov8n.pt')

# 2. Train the model
# Using absolute path to ensure the model can find the config file
dataset_path = os.path.abspath('data.yaml')

if not os.path.exists(dataset_path):
    raise FileNotFoundError(f'Could not find {dataset_path}. Please ensure the weapon_data_set folder exists.')

results = model.train(
    data=dataset_path, 
    epochs=100,
    imgsz=640,
    batch=16,
    name='weapon_yolo_model'
)
```

```python
# 3. Validate and Export
metrics = model.val()
print(f"Mean Average Precision: {metrics.box.map}")

# Export the trained weights to the models directory for use in detect.py
# path = model.export(format='onnx') 
```
